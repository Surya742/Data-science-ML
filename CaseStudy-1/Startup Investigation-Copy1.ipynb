```python
# What type of startups got funded more
startupType = df['IndustryVertical']
startupType
```

```python
startupType.dropna(inplace=True)
startupType
```

```python
freq = {}
for types in startupType:
    if types in freq:
        freq[types] += 1;
    else:
        freq[types] = 1
freq
```

```python
maxType = 0;
name = []
nameFreq = []
for ele in freq:
    name.append(ele)
    nameFreq.append(freq[ele])
    if freq[ele] > maxType:
        maxType = freq[ele]
maxType
# Hence "Consumer Internet" is the type of startup funded more
```
